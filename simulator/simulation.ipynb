```python
import ikpy.chain as ikchain
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# The URDF file is the mecanical description of the robot
urdf_file_path = "ergo_jr/ergo_jr.urdf"

# Goal coordinates
x = 0.1
y = 0.08
z = 0.1

# Load the chain from the URDF file and get the motor angles to get to this position
ergo_jr_chain = ikchain.Chain.from_urdf_file('ergo_jr/ergo_jr.urdf')
angles = ergo_jr_chain.inverse_kinematics([x, y, z])

# Plot the result
ax = plt.figure().add_subplot(111, projection='3d')
ergo_jr_chain.plot(angles, ax)
plt.show()
```
